```python
def add_prop(group):
    
    bir = group.births.astype(float)
    
    group['prop'] = bir / bir.sum()
    return group

names = names.groupby(['year', 'sex']).apply(add_prop)
```

```python
def get_qtd_count(group, q=0.5):
    g = group.sort_index(by='prop', ascending=False)
    return g.prop.cumsum().searchsorted(q)+1
```
